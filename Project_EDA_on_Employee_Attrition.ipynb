## Age

```python
#checking attrition rate for each age
table=pd.crosstab(attrition.Age, attrition.Attrition)
table['attrition percentage']=100*table[1]/(table[0]+table[1])
table
```

### Here we can see, that higher attrition rate is strongly associated with lower age group.

## BusinessTravel

```python
table=pd.crosstab(attrition.BusinessTravel, attrition.Attrition)
table['Attrition percentage']=100*table[1]/(table[0]+table[1])
table
```

```python
table=pd.crosstab(attrition.BusinessTravel, attrition.Attrition)
table['attrition percentage']=100*table[1]/(table[0]+table[1])

#Numbers of pairs of bars you want
N = len(table[0])

# Data on X-axis

# Specify the values of blue bars (height)
blue_bar = table[1]
# Specify the values of orange bars (height)
orange_bar = table[0]

# Position of bars on x-axis
ind = np.arange(N)

# Figure size
plt.figure(figsize=(10,5))

# Width of a bar 
width = 0.3       

# Plotting
plt.bar(ind, blue_bar , width, label='Attrition')
plt.bar(ind + width, orange_bar, width, label='Retention')

plt.xlabel('BusinessTravel')
plt.ylabel('Retention/Attrition Numbers')
plt.title('Attrition/Retention vs BusinessTravel')

# xticks()
# First argument - A list of positions at which ticks should be placed
# Second argument -  A list of labels to place at the given locations
plt.xticks(ind + width / 2, (table.index))

# Finding the best position for legends and putting it
plt.legend(loc='best')
plt.show()
```

### It's obvious from the crosstab that the attrition rate among employees in our company who travel frequently is comparatively higher than those who travel less or don't travel at all

```python
# Bivariate analysis between total working years and overtime
table = pd.crosstab(attrition.TotalWorkingYears,attrition.OverTime)
table['Percent_Employee_Overtime'] = table['Yes']*100/(table['No']+table['Yes'])
table
```

#### Employees in our organization are being subjected to over time irrespective of their work experience. Employees with higher work experience are more likely to be among the top level management and people with lesser work experience probably form the bottom core of the organization(subjective to the matter of qualification and skillset).

```python
# Multivariate analysis between TotalWorkingYears,OverTime and Attrition rate
# Given the work experience and whether the employee is subjected to overtime or not, 
# what is the probaility that he/she is likely to leave or stay in the organization?
table = pd.crosstab([attrition.TotalWorkingYears,attrition.OverTime],attrition['Attrition'],normalize='index')*100
table
```

#### From the above analysis it can be observed there's a strong negative association between the total work experience and overtime in our company with respect to atttrition rate. 
#### The attrition rate among employees who've been subjected to overtime with lesser work experience in our organization is quite high. 
#### Though employees having higher work experience are also being subjected to overtime, the attrition rate among these employees is comparitively less. People with lesser work experience are probably younger and are at the beginning stages of their career. Unable to cope up with the work pressure due to overtime, they are more likely to switch jobs.

```python
# Bivariate analysis between Job Level and OverTime
table = pd.crosstab(attrition.JobLevel,attrition.OverTime)
table['Percent_Employee_Overtime'] = table['Yes']*100/(table['No']+table['Yes'])
table
```

#### Considering the above analysis the percentage of people being subjected to overtime across all job levels seems to be similar. And it's obvious that number of people working at the lower job levels is comparitively higher than those working at higher job levels.
